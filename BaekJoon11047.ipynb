# 백준 11047번

```python
n, k = map(int,input().split())

coins = []
count = 0

for i in range(n):
    coins.append(int(input()))
    
for i in range(n-1,-1,-1):
    if k//coins[i] == 0:
        continue
    else: 
        count += k//coins[i]
        k %= coins[i]
    
print(count)   
```
