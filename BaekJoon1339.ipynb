```python
N = int(input())
a = []
dic = {'A':0, 'B':0, 'C':0,'D':0,'E':0,'F':0,'G':0,'H':0,'I':0,'J':0,
      'K':0,'L':0,'M':0,'N':0,'O':0,'P':0,'Q':0,'R':0,'S':0,'T':0,'U':0,
       'V':0,'W':0,'X':0,'Y':0,'Z':0}
value = {'A':0, 'B':0, 'C':0,'D':0,'E':0,'F':0,'G':0,'H':0,'I':0,'J':0,
      'K':0,'L':0,'M':0,'N':0,'O':0,'P':0,'Q':0,'R':0,'S':0,'T':0,'U':0,
       'V':0,'W':0,'X':0,'Y':0,'Z':0}

for i in range(N):
    a.append(input())


for i in range(N):
    c = len(a[i])
    for j in range(len(a[i])):
        dic[a[i][j]] += 10**c
        c -= 1

b = list(zip(dic.keys(),dic.values()))
b.sort(key = lambda x:x[1],reverse=True)


v = 9
for i in range(10):
    value[b[i][0]] = v
    v -= 1

sorted(value.items(), key = lambda item: item[1])
answer = 0

for i in range(N):
    c = len(a[i])
    for j in range(len(a[i])):
        answer += value[a[i][j]]*(10**(c-1))
        c -= 1
        

print(answer)
```
