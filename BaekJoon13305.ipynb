# 백준 13305번

```python
n = int(input())
km = list(map(int,input().split()))
price = list(map(int,input().split()))
price1 = price.copy()
del price1[len(price1)-1]

total = 0
mid = 0

for i,kms in enumerate(km):
    if price1[i] == min(price1):
        for j in range(i,len(km)):
            mid += km[j]
        
        total += mid*min(price1)
        break
        
    else:
        total += kms*price1[i]
        
    
print(total)
    
        
```
