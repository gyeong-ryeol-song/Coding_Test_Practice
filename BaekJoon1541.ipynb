# 백준 1541번

```python
note = input()

note = note.replace("+",".100000.")
note = note.replace("-",".100001.") 
note = note.split(".")

total = 0
mid = 0
temp = 0

for i in note:
    
    if temp == 0:
        if i != "100000" and i != "100001":
            total += int(i)
    if temp == 1:
        if i != "100000" and i != "100001":
            mid += int(i)
    if i == "100001":
        temp = 1
        total -= mid
        mid = 0

total -= mid    

print(total)
```
